```python
file_p = 'dataset.csv'
data = pd.read_csv(file_p)
data
```

```python
clf = OneClassSVM(kernel='linear', nu=0.5) # nu = 0.5 means we are allowing the svm model to have about 50% of the data as outliears
clf.fit(data)
```

```python
new_p = 'output.csv'
output_data = pd.read_csv(new_p)
output_data
```

```python
predictions = clf.predict(output_data)

output_data['prediction'] = predictions
```
